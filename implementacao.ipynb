```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('processamento_de_dados/resultado.csv')
data['volume corpo'] = np.array(data['comprimento corpo']) * np.pi*(np.array(data['largura corpo'])/2)**2
data['volume flagelo'] = np.array(data['comprimento flagelo']) * np.pi*(np.array(data['largura flagelo'])/2)**2
```

### Implementação

```python
import sklearn as sk
from sklearn import neighbors 
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix

#carregando dados
new_data = pd.read_csv('processamento_de_dados/resultado.csv')
#print(new_data)
y = new_data.iloc[:,4].values
#print(y)

x = new_data.iloc[:,[2,3]].values

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30)
```

```python
n=10
#treinando modelo 
knn= neighbors.KNeighborsClassifier(n,'uniform')
knn.fit(x_train,y_train)
y_pred= knn.predict(x_test)



cm = confusion_matrix(y_test,y_pred)
print(cm)
print("_______")
score = knn.score(x_test,y_test)
print(score)
```

```python
from sklearn.svm import SVC

#criando modelo SVC 
svc= SVC(kernel='rbf',gamma=0.1)

svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)

cm2 = confusion_matrix(y_test,y_pred)
print(cm2)
print("_______")
score2=svc.score(x_test,y_test)
print(score2)
```
