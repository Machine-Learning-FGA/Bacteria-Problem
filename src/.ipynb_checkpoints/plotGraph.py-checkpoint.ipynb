```python
import matplotlib.pyplot as pyplot
import numpy as np
import pandas as pd

comprimento_corpo = []
comprimento_flagelo = []
largura_corpo = []
largura_flagelo = []
tipo_bacteria = []
data_tsv = pd.read_csv('./data/data_cleaned.tsv', sep='\t')
rows = data_tsv.shape[0]

for i in range(0, rows):
    comprimento_corpo.append(data_tsv['comprimento corpo'][i])
    comprimento_flagelo.append(data_tsv['comprimento flagelo'][i])
    largura_corpo.append(data_tsv['largura corpo'][i])
    largura_flagelo.append(data_tsv['largura flagelo'][i])
    tipo_bacteria.append(data_tsv['Tipo'][i])
    
pyplot.scatter(tipo_bacteria, comprimento_corpo, color='blue')
pyplot.ylabel('Comprimento corpo')
pyplot.xlabel('tipo')
pyplot.title('Comprimento corpo x Tipo')
pyplot.grid(True)
pyplot.savefig('graph/comprimento_corpoxtipo.png')
pyplot.show()

pyplot.scatter(tipo_bacteria, comprimento_flagelo, color='red')
pyplot.ylabel('Comprimento flagelo')
pyplot.xlabel('tipo')
pyplot.title('Comprimento flagelo x Tipo')
pyplot.grid(True)
pyplot.savefig('graph/comprimento_flageloxtipo.png')
pyplot.show()

pyplot.scatter(tipo_bacteria, largura_corpo, color='green')
pyplot.ylabel('Largura corpo')
pyplot.xlabel('tipo')
pyplot.title('Largura corpo x Tipo')
pyplot.grid(True)
pyplot.savefig('graph/largura_corpoxtipo.png')
pyplot.show()

pyplot.scatter(tipo_bacteria, largura_flagelo, color='yellow')
pyplot.ylabel('Largura flagelo')
pyplot.xlabel('tipo')
pyplot.title('Largura flagelo x Tipo')
pyplot.grid(True)
pyplot.savefig('graph/larguraflageloxtipo.png')
pyplot.show()

area_corpo = [i*j for i,j in zip(comprimento_corpo,largura_corpo)]
area_flagelo = [i*j for i,j in zip(comprimento_flagelo,largura_flagelo)]
proporcao = [i/j for i,j in zip(area_flagelo, area_corpo)]
pyplot.scatter(tipo_bacteria, proporcao, color='red')
pyplot.ylabel('Proporcao(corpo/flagelo)')
pyplot.xlabel('tipo')
pyplot.title('Proporcao(corpo/flagelo) x Tipo')
pyplot.grid(True)
pyplot.savefig('graph/proporcaoxtipo.png')
pyplot.show()

```
