```python
# Importing Dependecies

import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
# Reading data from this file

data = pd.read_csv('dataset_bacteria.csv')
```

```python
# Creating new features for future analisys

data['volume corpo'] = np.array(data['comprimentocorpo']) * np.pi*(np.array(data['larguracorpo'])/2)**2
data['volume flagelo'] = np.array(data['comprimentoflagelo']) * np.pi*(np.array(data['larguraflagelo'])/2)**2
```

### Ploting Comprimento Corpo vs Largura Corpo Graph

```python
col1 = np.array(data['comprimentocorpo'])
col2 = np.array(data['larguracorpo'])

map = {1:'yellow', 2:'green', 3: 'red'}

for i in range(col1.size):
    mp.scatter(col1[i], col2[i], c=map[data['Tipo'][i]])
    
mp.show()    
```

### Ploting Comprimento Flagelo vs Largura Flagelo Graph

#### One of the best graphs found

```python
col1 = np.array(data['comprimentoflagelo'])
col2 = np.array(data['larguraflagelo'])

map = {1:'yellow', 2:'green', 3:'red'}

for i in range(col1.size):
    mp.scatter(col1[i], col2[i], c=map[data['Tipo'][i]])

mp.show()
```
